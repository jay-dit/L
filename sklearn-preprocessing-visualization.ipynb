### Разбить с сохранением пропорций классов

```python
# Создаем пропорции в отдельный столбец 
df['income_cat'] = pd.cut(df['median_income'],
                         bins = [0., 1.5, 3.0, 4.5, 6., np.inf],
                         labels=[1,2,3,4,5])
```

```python
split = StratifiedShuffleSplit(n_splits = 1,
                              test_size = 0.2,
                              random_state = 12)
```

```python
for train, test in split.split(df,df['income_cat']):
    strat_train_set = df.loc[train]
    strat_test_set = df.loc[test]
```
