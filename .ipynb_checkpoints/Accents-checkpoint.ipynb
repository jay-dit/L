```python
import pandas as pd
```

```python
with open('all_accents.txt', encoding = 'utf-8') as file1:
    text = file1.read()
textsp = text.split('\n')
textsp = list(map(lambda x: x.split('\t'), textsp))
df = pd.DataFrame(textsp)
```

```python
def find(name):
    k = list(df[1][df[0].__eq__(f'{name}')])
    k.append('')
    k = k[0]
    return ''.join(k)
```

```python
def rep():
    print('Введите слово, для проверки ударения или STOP')
    string = '!'
    while string != 'stop':
        string = str(input()).lower()
        print(find(string))
        print()
```

```python
rep()
```
