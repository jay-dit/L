```python
import numpy as np 
y = lambda x: np.cos(x)
dy = lambda x: -np.sin(x)
Nu = 0.1
n1 = -0.5
c = []
for j in range(200):
    n1 = n1 - Nu*dy(n1)
    c.append(n1)
c1 = np.array(c)
c2 = y(c1)
a = np.arange(-4, 2, 0.01)
b = y(a)
import matplotlib.pyplot as plt
plt.plot(a,b, label = 'sin(x)')
plt.scatter(c1, c2, c = 'orange', label = 'LR')
plt.legend(loc = 'lower right')
plt.show()
print('\nТочка минимума: ',n1,'\n\nМинимум:',y(n1))
```
